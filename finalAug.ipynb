```python
# ===================== 1. 라이브러리 임포트 및 환경 설정 =====================
import os
import cv2
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, Input, Model
import matplotlib.pyplot as plt
import random
from sklearn.metrics import classification_report
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# 환경 설정
BATCH_SIZE = 32
TARGET_SHAPE = (26, 26)         # (높이, 너비)
NUM_CHANNELS = 3                # 3채널 (목표: 검정, 초록, 빨간색)
NUM_CLASSES = 9
LEARNING_RATE = 1e-3
INITIAL_EPOCHS = 10             # 분류 모델 초기 학습 에포크 수
# 증강 대상 클래스 당 생성할 샘플 수를 1000개로 설정 (모든 클래스, 'none' 포함)
AUGMENT_SAMPLES_PER_CLASS = 1000  
EPOCHS_AE = 15                  # Autoencoder 학습 에포크
FIXMATCH_EPOCHS = 10            # (선택 사항) FixMatch 추가 학습 에포크
RANDOM_STATE = 42

np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)
random.seed(RANDOM_STATE)

# ===================== 2. 데이터 로딩 및 전처리 =====================
print("데이터 로딩 중...")
data = pd.read_pickle('LSWMD.pkl')
print("전체 데이터 shape:", data.shape)

# 라벨 데이터 (failureType 값이 있는 행)
data_labelled = data[data['failureType'].map(lambda x: list(x) != [])].copy()
print("라벨된 데이터 shape:", data_labelled.shape)

# 라벨 추출 (리스트 내부의 문자열) 및 컬럼명 수정
data_labelled['failureType'] = data_labelled['failureType'].map(lambda x: x[0][0])
data_labelled['trianTestLabel'] = data_labelled['trianTestLabel'].map(lambda x: x[0][0])
data_labelled.rename(columns={'trianTestLabel': 'trainTestLabel'}, inplace=True)
print("라벨 일부:")
print(data_labelled.head())

def process_wafer_map(wafer_map, target_size=TARGET_SHAPE):
    """
    waferMap (리스트)를 정규화된 3채널 이미지로 변환  
    색상 매핑: 
      0 -> [0,0,0] (검정), 
      1 -> [0,255,0] (녹색), 
      2 -> [255,0,0] (빨간색)
    최종 출력: (target_size[0], target_size[1], 3) 크기의 float32 이미지 (0~1 스케일)
    """
    img = np.array(wafer_map, dtype=np.uint8)
    img_resized = cv2.resize(img, (target_size[1], target_size[0]), interpolation=cv2.INTER_NEAREST)
    color_map = {0: [0, 0, 0],
                 1: [0, 255, 0],
                 2: [255, 0, 0]}
    colored_img = np.zeros((target_size[0], target_size[1], NUM_CHANNELS), dtype=np.uint8)
    for key, color in color_map.items():
        colored_img[img_resized == key] = color
    return colored_img.astype(np.float32) / 255.0

# 전처리 함수 테스트
example_img = process_wafer_map(data_labelled.iloc[0]['waferMap'])
plt.figure(figsize=(3,3))
plt.imshow(example_img)
plt.title(data_labelled.iloc[0]['failureType'])
plt.axis('off')
plt.show()

classes = ["Center", "Donut", "Edge-Loc", "Edge-Ring", "Loc", "Near-full", "Random", "Scratch", "none"]
print("클래스 목록:", classes)

# 훈련/테스트 데이터 분할 (각 클래스별 80:20 샘플링)
train_list = []
test_list = []
for cls in classes:
    df_cls = data_labelled[data_labelled['failureType'] == cls]
    df_train = df_cls.sample(frac=0.8, random_state=RANDOM_STATE)
    df_test = df_cls.drop(df_train.index)
    train_list.append(df_train)
    test_list.append(df_test)
    
train_df = pd.concat(train_list).reset_index(drop=True)
test_df = pd.concat(test_list).reset_index(drop=True)
print("훈련 데이터 shape:", train_df.shape)
print("테스트 데이터 shape:", test_df.shape)
print("훈련 데이터 클래스별 개수:")
print(train_df['failureType'].value_counts())
print("테스트 데이터 클래스별 개수:")
print(test_df['failureType'].value_counts())

class_to_idx = {name: i for i, name in enumerate(classes)}
print("클래스 to 인덱스 매핑:", class_to_idx)

def create_image_label_arrays(df):
    images = []
    labels = []
    for _, row in df.iterrows():
        img = process_wafer_map(row['waferMap'], target_size=TARGET_SHAPE)
        images.append(img)
        lbl = class_to_idx.get(row['failureType'], 0)
        labels.append(lbl)
    return np.stack(images), np.array(labels, dtype=np.int32)

X_train, y_train = create_image_label_arrays(train_df)
X_test, y_test = create_image_label_arrays(test_df)
print("X_train shape:", X_train.shape, "y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape, "y_test shape:", y_test.shape)

# 원본 훈련 샘플 이미지 시각화
plt.figure(figsize=(12, 3))
for i, cls in enumerate(classes):
    idxs = np.where(y_train == class_to_idx[cls])[0]
    if len(idxs) > 0:
        sample_idx = np.random.choice(idxs)
        plt.subplot(1, len(classes), i+1)
        plt.imshow(X_train[sample_idx])
        plt.title(cls)
        plt.axis('off')
plt.suptitle("원본 데이터: 각 클래스별 샘플 이미지")
plt.tight_layout()
plt.show()

# ===================== 3. 분류 모델 생성 및 초기 학습 =====================
def build_classifier_model(input_shape=(TARGET_SHAPE[0], TARGET_SHAPE[1], NUM_CHANNELS), num_classes=NUM_CLASSES):
    inputs = Input(shape=input_shape)
    x = layers.Conv2D(32, kernel_size=3, padding='same', activation='relu')(inputs)
    x = layers.MaxPooling2D(pool_size=3, padding='same')(x)
    x = layers.Conv2D(64, kernel_size=3, padding='same', activation='relu')(x)
    x = layers.MaxPooling2D(pool_size=3, padding='same')(x)
    x = layers.Conv2D(128, kernel_size=3, padding='same', activation='relu')(x)
    x = layers.GlobalMaxPooling2D()(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = Model(inputs, outputs)
    return model

classifier = build_classifier_model()
classifier.summary()

classifier.compile(optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
                     loss='sparse_categorical_crossentropy',
                     metrics=['accuracy'])

print("원본 데이터로 초기 학습 시작 (10 에포크)...")
history_initial = classifier.fit(X_train, y_train,
                                 batch_size=BATCH_SIZE,
                                 epochs=INITIAL_EPOCHS,
                                 validation_data=(X_test, y_test))

# 초기 모델 평가 (참고용)
initial_model = tf.keras.models.clone_model(classifier)
initial_model.set_weights(classifier.get_weights())
print("초기 모델 평가 (테스트 데이터):")
y_pred_initial = np.argmax(initial_model.predict(X_test), axis=-1)
print("초기 모델 Classification Report:")
print(classification_report(y_test, y_pred_initial, target_names=classes))

# ===================== 4. Autoencoder 학습 (증강용) =====================
# [수정된 부분] 각 클래스별로 최대 1000개씩 선택하여 균형 잡힌 Autoencoder 학습 데이터를 구성합니다.
X_train_ae_list = []
for cls in classes:
    cls_idx = class_to_idx[cls]
    idxs = np.where(y_train == cls_idx)[0]
    n_samples = len(idxs)
    n_use = min(1000, n_samples)  # 1000개 또는 해당 클래스 전체
    if n_use > 0:
        selected_idxs = np.random.choice(idxs, size=n_use, replace=False)
        X_train_ae_list.append(X_train[selected_idxs])
if len(X_train_ae_list) > 0:
    X_train_ae_balanced = np.concatenate(X_train_ae_list, axis=0)
else:
    X_train_ae_balanced = X_train
print("Balanced Autoencoder 학습 데이터 shape:", X_train_ae_balanced.shape)

# 각 이미지의 크기는 이미 (26,26,3)입니다.
# --- Autoencoder 구성 ---
input_tensor_ae = Input(shape=(TARGET_SHAPE[0], TARGET_SHAPE[1], NUM_CHANNELS))
# 인코더: Conv2D -> MaxPool2D (latent shape: (13,13,64))
x_ae = layers.Conv2D(64, (3,3), padding='same', activation='relu')(input_tensor_ae)
latent_vector = layers.MaxPool2D()(x_ae)
# 디코더: Conv2DTranspose -> UpSampling2D -> Conv2DTranspose
decode_layer_1 = layers.Conv2DTranspose(64, (3,3), padding='same', activation='relu')
decode_layer_2 = layers.UpSampling2D()    # (13,13,64) -> (26,26,64)
output_layer_ae = layers.Conv2DTranspose(3, (3,3), padding='same', activation='sigmoid')
x_dec = decode_layer_1(latent_vector)
x_dec = decode_layer_2(x_dec)
output_tensor_ae = output_layer_ae(x_dec)

ae = models.Model(inputs=input_tensor_ae, outputs=output_tensor_ae)
ae.compile(optimizer='adam', loss='mse')
ae.summary()

print("Autoencoder 학습 시작 (15 에포크)...")
ae.fit(X_train_ae_balanced, X_train_ae_balanced, batch_size=512, epochs=EPOCHS_AE, verbose=1)

# ===================== 5. 인코더 및 디코더 모델 추출 =====================
encoder = models.Model(inputs=input_tensor_ae, outputs=latent_vector)
latent_input = Input(shape=(13, 13, 64))
x_decoded = decode_layer_1(latent_input)
x_decoded = decode_layer_2(x_decoded)
decoded_output = output_layer_ae(x_decoded)
decoder = models.Model(inputs=latent_input, outputs=decoded_output)

# ===================== 후처리 함수: 연속적 3채널 출력 -> 이산적 3채널 이미지 =====================
def custom_discrete_postprocess_v2(img, black_thresh=0.1, green_delta=0.2):
    """
    img: (H, W, 3) 연속적 출력 (0~1)
    각 픽셀에 대해:
      - 평균 밝기가 black_thresh 미만이면 → 검정([0,0,0])
      - 그렇지 않고, 그린 채널이 최대이며 차이가 green_delta 이상이면 → 초록([0,1,0])
      - 그 외 → 빨강([1,0,0])
    """
    h, w, _ = img.shape
    out = np.empty((h, w, 3), dtype=np.float32)
    for i in range(h):
        for j in range(w):
            pixel = img[i, j, :]
            if np.mean(pixel) < black_thresh:
                out[i, j, :] = np.array([0, 0, 0])
            elif pixel[1] >= pixel[0] and pixel[1] >= pixel[2] and ((pixel[1] - pixel[0]) >= green_delta or (pixel[1] - pixel[2]) >= green_delta):
                out[i, j, :] = np.array([0, 1, 0])
            else:
                out[i, j, :] = np.array([1, 0, 0])
    return out

# ===================== 추가: Autoencoder 증강 전후 비교 시각화 =====================
aug_classes = [cls for cls in classes]  # 모든 클래스 (none 포함)
plt.figure(figsize=(2 * len(aug_classes), 4))
for i, cls in enumerate(aug_classes):
    cls_idx = class_to_idx[cls]
    idxs = np.where(y_train == cls_idx)[0]
    if len(idxs) == 0:
        continue
    sample_idx = np.random.choice(idxs)
    orig_img = X_train[sample_idx]  # (26,26,3)
    
    # 인코딩 → 노이즈 추가 → 디코딩 → 후처리
    latent = encoder.predict(np.expand_dims(orig_img, axis=0))
    latent_noisy = latent + np.random.normal(loc=0, scale=0.2, size=latent.shape)
    aug_img_cont = decoder.predict(latent_noisy)[0]  # 연속적 출력
    aug_img = custom_discrete_postprocess_v2(aug_img_cont, black_thresh=0.1, green_delta=0.2)
    
    plt.subplot(2, len(aug_classes), i+1)
    plt.imshow(orig_img)
    plt.title(f"{cls}\nOriginal")
    plt.axis('off')
    
    plt.subplot(2, len(aug_classes), len(aug_classes)+i+1)
    plt.imshow(aug_img)
    plt.title(f"{cls}\nAugmented")
    plt.axis('off')
plt.suptitle("Autoencoder 증강 전후 비교 (각 클래스별)", fontsize=16)
plt.tight_layout()
plt.show()

# ===================== 6. 증강 이미지 생성 (벡터화 방식) =====================
def augment_class_data_with_trained_ae(X_cls, n_augment, encoder, decoder, base_noise_std=0.15):
    """
    X_cls: 원본 이미지 배열 (각 이미지 크기: 26x26x3)
    → X_cls에서 무작위 n_augment 샘플 선택 후 encoder → 노이즈 추가 → decoder를 통해 증강 이미지 생성.
    후처리로 custom_discrete_postprocess_v2 적용.
    """
    latent_vals = encoder.predict(X_cls, batch_size=512, verbose=0)  # (n_samples, 13,13,64)
    indices = np.random.randint(0, latent_vals.shape[0], size=n_augment)
    selected_latents = latent_vals[indices]
    noise = np.random.normal(loc=0, scale=base_noise_std, size=selected_latents.shape)
    latent_noisy = selected_latents + noise
    augmented_cont = decoder.predict(latent_noisy, batch_size=512, verbose=0)
    augmented_images = np.array([custom_discrete_postprocess_v2(img, black_thresh=0.1, green_delta=0.2)
                                   for img in augmented_cont])
    return augmented_images

# 각 클래스별 증강 목표: 모든 클래스(‘none’ 포함)에서 1000개씩 생성
augment_counts = {}
for cls in classes:
    augment_counts[cls] = AUGMENT_SAMPLES_PER_CLASS
print("각 클래스별 추가 증강 목표 수:", augment_counts)

synthetic_images_list = []
synthetic_labels_list = []
for cls in classes:
    cls_idx = class_to_idx[cls]
    idxs = np.where(y_train == cls_idx)[0]
    X_cls = X_train[idxs]  # (n,26,26,3)
    n_samples = X_cls.shape[0]
    n_augment = augment_counts[cls]
    print(f"Class {cls} (index {cls_idx}) - 원본 샘플 수: {n_samples}, 증강 목표: {n_augment}")
    if n_augment > 0:
        X_aug = augment_class_data_with_trained_ae(X_cls, n_augment, encoder, decoder)
        synthetic_images_list.append(X_aug)
        synthetic_labels_list.append(np.full((X_aug.shape[0],), cls_idx))
        print(f"Class {cls} - 생성된 증강 샘플 수: {X_aug.shape[0]}")
if len(synthetic_images_list) > 0:
    synthetic_images = np.concatenate(synthetic_images_list, axis=0)
    synthetic_labels = np.concatenate(synthetic_labels_list, axis=0)
else:
    synthetic_images = np.empty((0, TARGET_SHAPE[0], TARGET_SHAPE[1], NUM_CHANNELS))
    synthetic_labels = np.empty((0,), dtype=np.int32)
print("전체 증강 이미지 shape:", synthetic_images.shape)
print("전체 증강 레이블 shape:", synthetic_labels.shape)

# 증강 이미지 시각화 (각 클래스별 2개 샘플)
aug_classes = [cls for cls in classes]
n_cols = len(aug_classes)
plt.figure(figsize=(3 * n_cols, 6))
for i, cls in enumerate(aug_classes):
    cls_idx = class_to_idx[cls]
    idxs = np.where(synthetic_labels == cls_idx)[0]
    if len(idxs) > 0:
        sample_idxs = np.random.choice(idxs, size=2, replace=False) if len(idxs) >= 2 else [idxs[0], idxs[0]]
        plt.subplot(2, n_cols, i+1)
        plt.imshow(synthetic_images[sample_idxs[0]])
        plt.title(cls)
        plt.axis('off')
        plt.subplot(2, n_cols, n_cols + i + 1)
        plt.imshow(synthetic_images[sample_idxs[1]])
        plt.axis('off')
plt.suptitle("증강 데이터: 각 클래스별 2개 샘플 이미지", fontsize=16)
plt.tight_layout()
plt.show()

# ===================== 7. 원본 데이터 + 증강 데이터 합치기 및 새 모델 학습 =====================
# 원본 데이터는 모두 사용하고, 각 클래스별 증강 데이터 1000개씩을 합칩니다.
X_train_aug = np.concatenate([X_train, synthetic_images], axis=0)
y_train_aug = np.concatenate([y_train, synthetic_labels], axis=0)
print("원본 + 증강 데이터 합친 후 X_train shape:", X_train_aug.shape, "y_train shape:", y_train_aug.shape)

# 데이터를 섞어줍니다.
shuffle_idx = np.random.permutation(X_train_aug.shape[0])
X_train_aug = X_train_aug[shuffle_idx]
y_train_aug = y_train_aug[shuffle_idx]

# 새 모델 생성 (처음부터 학습)
new_classifier = build_classifier_model()
new_classifier.compile(optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
                         loss='sparse_categorical_crossentropy',
                         metrics=['accuracy'])

print("새로운 모델 학습 시작 (10 에포크)...")
history_new = new_classifier.fit(X_train_aug, y_train_aug,
                                 batch_size=BATCH_SIZE,
                                 epochs=INITIAL_EPOCHS,
                                 validation_data=(X_test, y_test))

print("\n[새로운 모델 평가 (테스트 데이터)]")
y_pred_new = np.argmax(new_classifier.predict(X_test), axis=-1)
print("새로운 모델 Classification Report:")
print(classification_report(y_test, y_pred_new, target_names=classes))

```

```python

# ===================== 8. FixMatch 추가 학습 =====================
# FixMatch 하이퍼파라미터
CONFIDENCE_THRESHOLD = 0.80
UNSUPERVISED_LOSS_WEIGHT = 1.0
FIXMATCH_EPOCHS = 10

# (a) Unlabeled 데이터용 augmentation 함수 정의
def weak_augment_unlabeled(image):
    """
    약한 augmentation: 좌우 반전, 패딩 후 랜덤 크롭 등 간단한 변환.
    """
    image = tf.convert_to_tensor(image)
    image = tf.image.random_flip_left_right(image)
    image = tf.image.resize_with_crop_or_pad(image, TARGET_SHAPE[0] + 4, TARGET_SHAPE[1] + 4)
    image = tf.image.random_crop(image, size=[TARGET_SHAPE[0], TARGET_SHAPE[1], NUM_CHANNELS])
    return image

def strong_augment_unlabeled(image):
    """
    강한 augmentation: Autoencoder의 encoder/decoder를 이용하여,
    latent 공간에 노이즈를 추가한 후 디코딩하여 생성.
    """
    image = tf.convert_to_tensor(image)
    image_batch = tf.expand_dims(image, axis=0)  # shape (1, H, W, C)
    latent = encoder(image_batch, training=False)
    noise = tf.random.normal(shape=tf.shape(latent), mean=0.0, stddev=0.1)
    latent_noisy = latent + noise
    augmented_image = decoder(latent_noisy, training=False)
    augmented_image = tf.squeeze(augmented_image, axis=0)
    return augmented_image

# (b) unlabeled 데이터셋 구성  
# 여기서는 새로 추출한 X_unlabeled를 사용합니다.
labeled_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train)).shuffle(1000, seed=RANDOM_STATE).batch(BATCH_SIZE)
unlabeled_ds = tf.data.Dataset.from_tensor_slices(X_unlabeled).shuffle(1000, seed=RANDOM_STATE).batch(BATCH_SIZE)

# FixMatch 전용 옵티마이저 (필요에 따라 학습률 등을 조정)
fixmatch_optimizer = optimizers.Adam(learning_rate=LEARNING_RATE)

@tf.function
def fixmatch_train_step(x_l, y_l, x_ul):
    # unlabeled 데이터에 weak 및 strong augmentation 적용
    x_ul_weak = tf.map_fn(weak_augment_unlabeled, x_ul, fn_output_signature=tf.float32)
    x_ul_strong = tf.map_fn(strong_augment_unlabeled, x_ul, fn_output_signature=tf.float32)
    
    with tf.GradientTape() as tape:
        # labeled 데이터에 대한 supervised loss
        logits_l = classifier(x_l, training=True)
        sup_loss = tf.keras.losses.sparse_categorical_crossentropy(y_l, logits_l)
        sup_loss = tf.reduce_mean(sup_loss)
        
        # unlabeled 데이터: weak augmented 결과로 pseudo-label 생성
        logits_ul_weak = classifier(x_ul_weak, training=True)
        probs_ul_weak = tf.nn.softmax(logits_ul_weak, axis=-1)
        max_probs = tf.reduce_max(probs_ul_weak, axis=-1)
        pseudo_labels = tf.argmax(probs_ul_weak, axis=-1, output_type=tf.int32)
        mask = tf.cast(max_probs >= CONFIDENCE_THRESHOLD, tf.float32)
        
        # unlabeled 데이터: strong augmented 결과에 대해 unsupervised loss 계산
        logits_ul_strong = classifier(x_ul_strong, training=True)
        unsup_loss = tf.keras.losses.sparse_categorical_crossentropy(pseudo_labels, logits_ul_strong)
        unsup_loss = tf.reduce_mean(unsup_loss * mask)
        
        total_loss = sup_loss + UNSUPERVISED_LOSS_WEIGHT * unsup_loss
    
    gradients = tape.gradient(total_loss, classifier.trainable_variables)
    fixmatch_optimizer.apply_gradients(zip(gradients, classifier.trainable_variables))
    
    return total_loss, sup_loss, unsup_loss, tf.reduce_mean(mask)

# (c) FixMatch 학습 루프
fixmatch_train_history = {'total_loss': [], 'sup_loss': [], 'unsup_loss': [], 'mask_mean': []}
for epoch in range(FIXMATCH_EPOCHS):
    epoch_total_loss = 0.0
    epoch_sup_loss = 0.0
    epoch_unsup_loss = 0.0
    epoch_mask_mean = 0.0
    steps = 0
    # labeled와 unlabeled 배치를 동시에 사용 (두 데이터셋의 크기가 다를 수 있으므로, 짧은 쪽에 맞춤)
    for (x_l_batch, y_l_batch), x_ul_batch in zip(labeled_ds, unlabeled_ds):
        total_loss, sup_loss, unsup_loss, mask_mean = fixmatch_train_step(x_l_batch, y_l_batch, x_ul_batch)
        epoch_total_loss += total_loss
        epoch_sup_loss += sup_loss
        epoch_unsup_loss += unsup_loss
        epoch_mask_mean += mask_mean
        steps += 1
    avg_total_loss = epoch_total_loss / steps
    avg_sup_loss = epoch_sup_loss / steps
    avg_unsup_loss = epoch_unsup_loss / steps
    avg_mask_mean = epoch_mask_mean / steps
    fixmatch_train_history['total_loss'].append(avg_total_loss.numpy())
    fixmatch_train_history['sup_loss'].append(avg_sup_loss.numpy())
    fixmatch_train_history['unsup_loss'].append(avg_unsup_loss.numpy())
    fixmatch_train_history['mask_mean'].append(avg_mask_mean.numpy())
    print(f"FixMatch Epoch {epoch+1}/{FIXMATCH_EPOCHS}: total_loss={avg_total_loss.numpy():.4f}, sup_loss={avg_sup_loss.numpy():.4f}, unsup_loss={avg_unsup_loss.numpy():.4f}, mask_mean={avg_mask_mean.numpy():.4f}")

# (d) FixMatch 추가 학습 후 평가
print("\n[FixMatch 추가 학습 후 평가 (테스트 데이터)]")
evaluate_model(classifier, X_test, y_test, title="FixMatch 최종 모델")

# FixMatch 학습 곡선 시각화
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(fixmatch_train_history['total_loss'], label='Total Loss')
plt.plot(fixmatch_train_history['sup_loss'], label='Supervised Loss')
plt.plot(fixmatch_train_history['unsup_loss'], label='Unsupervised Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('FixMatch Training Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(fixmatch_train_history['mask_mean'], label='Mask Mean (Confidence Ratio)')
plt.xlabel('Epoch')
plt.ylabel('Ratio')
plt.title('FixMatch Mask Mean')
plt.legend()
plt.tight_layout()
plt.show()
```
